```python
import re
import numpy as np

# Function to extract features
def extract_features(url):
    features = {}
    features['url_length'] = len(url)  # URL Length
    features['count_https'] = 1 if 'https' in url else 0  # HTTPS Presence
    features['count_at'] = url.count('@')  # Count '@'
    features['count_hyphen'] = url.count('-')  # Count '-'
    features['count_dot'] = url.count('.')  # Count '.'
    features['count_slash'] = url.count('/')  # Count '/'
    features['count_digits'] = sum(c.isdigit() for c in url)  # Count digits

    # Check if URL has an IP address
    features['has_ip'] = 1 if re.match(r'\d+\.\d+\.\d+\.\d+', url) else 0

    # Check if URL has suspicious words
    suspicious_words = ['login', 'bank', 'secure', 'account', 'verify', 'password', 'update']
    features['has_suspicious_words'] = 1 if any(word in url for word in suspicious_words) else 0

    return features

# Apply feature extraction
features_df = df['URLs'].apply(lambda url: extract_features(url))
features_df = pd.DataFrame(features_df.tolist())

# Combine features with labels
final_df = pd.concat([features_df, df['Labels']], axis=1)

# Display the processed dataset
print(final_df.head())
```

```python
import pandas as pd
import numpy as np
import re
import tldextract
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report

# Load Data
file_path = "data_bal - 20000.xlsx"  # Update path if needed
df = pd.read_excel(file_path)

# Rename Columns for Consistency
df.columns = ["Label", "URL"]

# **🚀 Function to Tokenize URL**
def tokenize_url(url):
    url = str(url)
    extracted = tldextract.extract(url)
    
    subdomain = extracted.subdomain.replace('.', ' ')  # Tokenize subdomain
    domain = extracted.domain  # Main domain
    suffix = extracted.suffix  # .com, .net, .org etc.
    
    # Extract path
    path = re.sub(r'[^\w\s]', ' ', url.split('/')[-1])  # Remove special chars from path
    
    return f"{subdomain} {domain} {suffix} {path}"  # Combine tokens

# Apply URL Tokenization
df["tokenized_url"] = df["URL"].apply(tokenize_url)

# **🚀 TF-IDF Vectorization**
vectorizer = TfidfVectorizer(max_features=5000)
X_tfidf = vectorizer.fit_transform(df["tokenized_url"]).toarray()

# **Feature Engineering: Extract URL Features**
def extract_features(url):
    url = str(url)
    features = {}
    
    # URL Length
    features['url_length'] = len(url)

    # Number of dots in URL
    features['num_dots'] = url.count('.')

    # Number of hyphens
    features['num_hyphens'] = url.count('-')

    # Number of digits in URL
    features['num_digits'] = sum(c.isdigit() for c in url)

    # Presence of "@" symbol
    features['has_at_symbol'] = 1 if "@" in url else 0

    # Presence of "https"
    features['has_https'] = 1 if url.startswith("https") else 0

    # Check for suspicious words in URL
    suspicious_words = ["login", "secure", "verify", "banking", "account", "update", "free", "click"]
    features['has_suspicious_word'] = 1 if any(word in url.lower() for word in suspicious_words) else 0

    return features

# Apply Feature Extraction
features_df = df['URL'].apply(extract_features).apply(pd.Series)

# **Merge Tokenized TF-IDF and Extracted Features**
X_combined = np.hstack((X_tfidf, features_df.values))

# Define Labels
y = df['Label']

# **🚀 Train-Test Split**
X_train, X_test, y_train, y_test = train_test_split(X_combined, y, test_size=0.2, random_state=42)

# **Normalize Data**
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# **Train a Random Forest Model**
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# **Predictions**
y_pred = model.predict(X_test)

# **Model Evaluation**
accuracy = accuracy_score(y_test, y_pred)
print(f"🚀 Accuracy: {accuracy:.4f}")

# **Classification Report**
print("\n📊 Classification Report:\n", classification_report(y_test, y_pred))
```

```python
import numpy as np

def predict_url(url):
    """Preprocess and predict whether a URL is phishing or legitimate."""
    
    # Tokenize URL
    tokenized = tokenize_url(url)
    
    # Vectorize using the trained TF-IDF vectorizer
    vectorized = vectorizer.transform([tokenized]).toarray()
    
    # Extract URL-based numerical features
    features = extract_features(url)
    features_array = np.array([list(features.values())])
    
    # Combine TF-IDF and numerical features
    combined_features = np.hstack((vectorized, features_array))
    
    # Ensure the test sample has the same number of features as the training data
    if combined_features.shape[1] != X_train.shape[1]:
        raise ValueError(f"Feature mismatch: Expected {X_train.shape[1]}, but got {combined_features.shape[1]}")
    
    # Predict using the trained model
    prediction = model.predict(combined_features)
    
    return "Phishing" if prediction[0] == 1 else "Legit"

# **🔹 Test Predictions**
test_url1 = "http://paypal-login-update-secure.com"  # Example phishing URL
print(f"Prediction for {test_url1}: {predict_url(test_url1)}")

test_url2 = "https://www.paypal.com/signin"  # Example legitimate URL
print(f"Prediction for {test_url2}: {predict_url(test_url2)}")
```

```python
import re
from urllib.parse import urlparse

# Define suspicious keywords often found in phishing URLs
suspicious_keywords = ["secure", "login", "verify", "update", "account", "bank", "paypal", "confirm"]

# Function to analyze the URL for phishing indicators
def analyze_url(url):
    parsed_url = urlparse(url)
    domain = parsed_url.netloc.lower()

    # Extract features
    use_https = parsed_url.scheme == "https"
    num_dots = url.count(".")
    num_hyphens = url.count("-")
    num_special_chars = sum(url.count(c) for c in ["@", "?", "=", "&", "%"])
    has_ip = bool(re.match(r"^\d{1,3}(\.\d{1,3}){3}$", domain))
    url_length = len(url)
    keyword_flag = any(word in domain for word in suspicious_keywords)

    # Construct explanation
    explanation = f"\n🔍 **Analyzing URL:** {url}\n"

    if use_https:
        explanation += "✅ The URL uses HTTPS, which is a good sign. Secure websites typically encrypt your data to protect against attackers.\n"
    else:
        explanation += "⚠️ The URL does not use HTTPS, meaning it lacks basic encryption. This is a major red flag because legitimate websites almost always use HTTPS for security.\n"

    if num_dots > 2:
        explanation += f"⚠️ The domain has {num_dots} dots, which is unusually high. Phishing websites often add extra subdomains to imitate legitimate sites.\n"

    if num_hyphens > 1:
        explanation += f"⚠️ The URL contains {num_hyphens} hyphens. Attackers frequently use hyphens to create fake versions of real websites (e.g., 'paypal-login-secure.com' instead of 'paypal.com').\n"

    if num_special_chars > 2:
        explanation += f"⚠️ This URL has {num_special_chars} special characters like '@', '?', or '='. These are often used in phishing URLs to obfuscate the real domain or redirect users to another page.\n"

    if has_ip:
        explanation += "⚠️ The domain is actually an IP address instead of a regular website name. Phishing scams sometimes use IP addresses to avoid detection by security filters.\n"
    else:
        explanation += "✅ The URL contains a standard domain name, which is more common for legitimate websites.\n"

    if url_length > 75:
        explanation += f"⚠️ This URL is {url_length} characters long. Extremely long URLs are often used to hide malicious content or trick users into clicking suspicious links.\n"

    if keyword_flag:
        explanation += "⚠️ The domain contains words like 'login', 'secure', or 'verify'. These are frequently used in phishing attacks to make a fake site seem trustworthy.\n"

    # Final decision
    if not use_https or has_ip or num_dots > 2 or num_hyphens > 1 or num_special_chars > 2 or url_length > 75 or keyword_flag:
        explanation += "\n🚨 **Conclusion:** This URL has multiple red flags and is likely a phishing attempt. Be cautious before clicking or entering any personal information!\n"
    else:
        explanation += "\n✅ **Conclusion:** No major phishing indicators were found in this URL. However, always be cautious and verify before entering sensitive data.\n"

    print(explanation)

# List of test URLs
test_urls = [
    "http://paypal-login-update-secure.com",
    "https://www.amazon.com",
    "https://secure-verify-instagram.com",
    "https://login.microsoftonline.com",
    "https://www.netflix.com/login",
    "https://accounts.google.com/ServiceLogin",
    "https://www.microsoft.com/en-us",
    "http://facebook-verification-support.com",
    "http://microsoft-verification-support.com"
]

# Analyze each URL
for url in test_urls:
    analyze_url(url)
```
